```python
xc = data['Category'].unique()
xc
```

```python
#encoding category
category = {}
for i in range(0,len(xc)):
    category[xc[i]]=i
```

```python
#apply to the column
data['Category_encode'] = data['Category'].map(category) 
```

```python
data.head()
```

```python
# cleaning Size
xs = data['Size'].unique()
xs
```

```python
x3 = xs[1][:-1]
x3
```

```python
def size_change(x):
    if 'M' in x:
      
        return (float(x[:-1])*1000000)
    elif 'k' in x:
        return (float(x[:-1])*1000)
    else:
        return None
```

```python
data['Size'] = data['Size'].map(size_change)
```

```python
data.head()
```

```python
df2 = pd.get_dummies(data, columns=['Category'])
```

```python
df2
```

```python
data['Installs'] = [int(i[:-1].replace(',','')) for i in data['Installs']]
```

```python
data.head()
```

```python
def type_change(x):
    if x == 'Free':
        return 0
    else:
        return 1
data['Type'] = data['Type'].map(type_change)
```

```python
data.head()
```

```python
data['Content Rating'].unique()
```

```python
content = {}
for i in range(0, len(data['Content Rating'].unique())):
    content[data['Content Rating'].unique()[i]] = i
```

```python
data['Content Rating'] = data['Content Rating'].map(content)
```

```python
data.head()
```

```python

data['Price'].unique()
```

```python
# return the the price w/ $
def price_change(x):
    if x == '0':
        return 0
    else:
        x = x[1:]
        x = float(x)
        return x
```

```python
data['Price'] = data['Price'].map(price_change).astype(float)
```

```python
data.head()
```

```python
data['Reviews'].astype(int)
```

```python
xg = data['Genres'].unique()
```

```python
genres = {}
for i in range(len(xg)):
    genres[xg[i]] = i
```

```python
data['Genres_encoded'] = data['Genres'].map(genres).astype(int)
```

```python
# will be removed 



#for evaluation of error term and 
def Evaluationmatrix(y_true, y_predict):
    print ('Mean Squared Error: '+ str(metrics.mean_squared_error(y_true,y_predict)))
    print ('Mean absolute Error: '+ str(metrics.mean_absolute_error(y_true,y_predict)))
    print ('Mean squared Log Error: '+ str(metrics.mean_squared_log_error(y_true,y_predict)))
```

```python
# will be removed 


def Evaluationmatrix_dict(y_true, y_predict, name = 'Linear - Integer'):
    dict_matrix = {}
    dict_matrix['Series Name'] = name
    dict_matrix['Mean Squared Error'] = metrics.mean_squared_error(y_true,y_predict)
    dict_matrix['Mean Absolute Error'] = metrics.mean_absolute_error(y_true,y_predict)
    dict_matrix['Mean Squared Log Error'] = metrics.mean_squared_log_error(y_true,y_predict)
    return dict_matrix
```
